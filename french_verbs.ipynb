Source: https://thelanguagegarage.com/regular-french-er-verbs/

According to this article most French verbs are regular. And all -er verbs are regular. The only thing is that you have to keep in mind that the spelling changes for some of them in order to keep the correct pronounciation. E.g. for manger, the intuitive way if you´re following the "parler" model would be nous mangons - but then the pronounciation would sound more like the fruit, so we spell it "mangeons" - adding the e - so that it goes from harder g sound to a softer g/j sound. There´s more examples like this so you can check this article out when looking more closely at the -er verbs.
